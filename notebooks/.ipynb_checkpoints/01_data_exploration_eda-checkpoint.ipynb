# Step 3: Target Variable Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Parts_Per_Hour'], kde=True)
plt.title("Distribution of Parts Per Hour")
plt.show()
```

# Step 4: Feature vs Target

```python
sns.scatterplot(x='Cycle_Time', y='Parts_Per_Hour', data=df)
plt.title("Cycle Time vs Parts Per Hour")
plt.show()
```

# Step 5: Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
