# STEP 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import joblib
```

# STEP 2: Load Model & Feature Importance

```python
model = joblib.load("../models/linear_regression_model.pkl")
feature_cols = joblib.load("../models/feature_columns.pkl")
```

# STEP 3: Create Feature Importance Table

```python
coef_df = pd.DataFrame({
    "Feature": feature_cols,
    "Coefficient": model.coef_
})

coef_df["Abs_Impact"] = coef_df["Coefficient"].abs()
coef_df = coef_df.sort_values(by="Abs_Impact", ascending=False)

coef_df
```

# STEP 4: Key Positive Drivers of Output

Key Output-Boosting Factors
- Efficiency Score: Strongest positive contributor to machine output.
- Cooling Time: Proper cooling improves part quality and throughput.
- Injection Temperature & Pressure: Higher optimal values increase mold filling efficiency.
- Premium Material Grade: Leads to smoother flow and higher production rates.
- Operator Experience: Skilled operators significantly improve output.

# Key Negative Drivers of Output

```python
Key Output-Reducing Factors
- Total Cycle Time: Largest negative impact; longer cycles reduce hourly output.
- Machine Age: Older machines are less efficient and require frequent maintenance.
- Night Shift: Slightly lower output compared to day shifts.
- Material Viscosity: High viscosity materials reduce flow efficiency.
- Machine Type B/C: Lower performance compared to baseline machines.
```
